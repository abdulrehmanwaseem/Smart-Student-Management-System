# Smart Student Management System

```python
def validate_input(prompt, field_name, required=True, numeric=False):
    while True:
        value = input(prompt).strip()
        if not value and required:
            print(f"{field_name} cannot be empty. Please try again.")
            continue
        if numeric and value:
            if not value.isdigit():
                print(f"{field_name} must be a number. Please try again.")
                continue
        return value

def add_student():
    id = validate_input("Enter Student ID: ", "Student ID", numeric=True)
    name = validate_input("Enter Student Name: ", "Name")
    age = validate_input("Enter Student Age: ", "Age", numeric=True)
    grade = validate_input("Enter Student Grade: ", "Grade")
    marks = validate_input("Enter Student Marks: ", "Marks", numeric=True)

    with open("students.txt", "a") as file:
        file.write(f"{id},{name},{age},{grade},{marks}\n")

    print("Studnet added succesfully")


def view_students():
    print("=== Studnet List ===")
    with open("students.txt", "r") as file:
        students = file.readlines()
    for student in students:
        id, name, age, grade, marks = student.strip().split(",")
        print(f"ID: {id}, Name: {name}, Age: {age}, Grade: {grade}, Marks: {marks}")


def search_student():
    search_by = input("Serach by (1) ID or (2) Name: ")
    if search_by == "1":
        user_id = input("Enter studnet ID to serach: ")
        user_id = int(user_id)
    elif search_by == "2":
        user_name = input("Enter studnet Name to serach: ").strip().lower()
    else:
        print("Invalid option.")
        return

    with open("students.txt", "r") as file:
        students = file.readlines()

    for student in students:
        id, name, age, grade, marks = student.strip().split(",")
        if search_by == "2":
            if user_name in name.strip().lower():
                print("Studnet found")
                print(f"ID: {id}, Name: {name}, Age: {age}, Grade: {grade}, Marks: {marks}")
                return
        else:
            if int(id) == user_id:
                print("Studnet found")
                print(f"ID: {id}, Name: {name}, Age: {age}, Grade: {grade}, Marks: {marks}")
                return
    print("Studnet not found")


def update_studnet():
    user_id = input("Enter studnet ID to update: ")
    user_id = int(user_id)

    with open("students.txt", "r") as file:
        students = file.readlines()

    for i, student in enumerate(students):
        id, name, age, grade, marks = student.strip().split(",")
        if int(id) == user_id:
            print("Studnet found")
            print(f"ID: {id}, Name: {name}, Age: {age}, Grade: {grade}, Marks: {marks}")

            new_name = input("Enter new name (leave blank to keep curent): ")
            new_age = input("Enter new age (leave blank to keep curent): ")
            new_grade = input("Enter new grade (leave blank to keep curent): ")
            new_marks = input("Enter new marks (leave blank to keep curent): ")

            if new_name:
                name = new_name
            if new_age:
                age = new_age
            if new_grade:
                grade = new_grade
            if new_marks:
                marks = new_marks

            students[i] = f"{id},{name},{age},{grade},{marks}\n"

            with open("students.txt", "w") as file:
                file.writelines(students)

            print("Studnet record updated succesfully!")
            return
    print("Student not found")


def analyze_data():
    print("=== Data Analys ===")
    with open("students.txt", "r") as file:
        students = file.readlines()

    marks = [int(s.strip().split(",")[-1]) for s in students]

    average_marks = sum(marks) / len(marks)
    highest_marks = max(marks)
    lowest_marks = min(marks)

    top_student = max(students, key=lambda s: int(s.strip().split(",")[-1]))
    top_name, top_marks = top_student.strip().split(",")[1], int(top_student.strip().split(",")[-1])

    below_avg_count = sum(
        1
        for s in students
        if int(s.strip().split(",")[-1]) < average_marks
    )

    print(f"Average Marks: {average_marks:.2f}")
    print(f"Top Perfomer: {top_name} ({top_marks})")
    print(f"Studnets Below Average: {below_avg_count}")
    print(f"Highest Marks: {highest_marks} | Lowest: {lowest_marks}")


def main():
    while True:
        print("\n=== Studnet Managment System ===")
        print("1. Add Studnet")
        print("2. View Studnets")
        print("3. Serach Studnet")
        print("4. Update Studnet")
        print("5. Analize Data")
        print("6. Exit")

        choice = input("Enter your choise: ")

        if choice == "1":
            add_student()
        elif choice == "2":
            view_students()
        elif choice == "3":
            search_student()
        elif choice == "4":
            update_studnet()
        elif choice == "5":
            analyze_data()
        elif choice == "6":
            print("Exiting...")
            break
        else:
            print("Invalied choise.")

main()
```
